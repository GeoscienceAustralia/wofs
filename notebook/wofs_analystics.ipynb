```python
import rasterio
import sys,os,time

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def get_water_summary(wsumfile,cellid,year):
    rasfile = rasterio.open(wsumfile)
    
    waterob=rasfile.read(1)[:,:]
    clearob=rasfile.read(2)[:,:]
    
    rasfile = None
    
    return (waterob, clearob)
    
def get_uniq_waterobs_pixels_mask(waterob):
    #print "Number of Unique Water Pixels= ", np.sum(waterob>0)
    return waterob>0

def get_uniq_clearobs_pixels_mask(clearob):
    #print "Number of ClearObserved Pixels= ", np.sum(clearob>0)
    return clearob>0
    
def images2display(img1,img2):
    """Display 2 images side by side
    """
    f, ax = plt.subplots(1,2,figsize=(20,10))
    ax[0].imshow(img1, cmap='Greys')
    ax[0].set_title('%s\n'%('Image1'))
    #plt.show()    

    ax[1].imshow(img2, cmap='Greys')
    ax[1].set_title('%s\n'%('Image2')) 

    plt.show()
```
